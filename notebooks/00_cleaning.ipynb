### 2 - Load Raw Data

```python
# Load the raw dataset
AC = pd.read_csv("../data/raw/application_categories.csv")
AD = pd.read_csv("../data/raw/application_developers.csv")
AG = pd.read_csv("../data/raw/application_genres.csv")
AP = pd.read_csv("../data/raw/application_platforms.csv")
APu = pd.read_csv("../data/raw/application_publishers.csv")
application = pd.read_csv("../data/raw/applications.csv")
```

#### the AI suggested multiple ways to load the datasets on of them is to ot put each dataset in a different variable
#### but I think it is better to put all the datasets in one variable and then we can use it to clean the data and 
#### then we can use it to analyze the data.


```python
# Load additional datasets
categories = pd.read_csv("../data/raw/categories.csv")
developers = pd.read_csv("../data/raw/developers.csv")
genres = pd.read_csv("../data/raw/genres.csv")
platforms = pd.read_csv("../data/raw/platforms.csv")
publishers = pd.read_csv("../data/raw/publishers.csv")
reviews = pd.read_csv("../data/raw/reviews.csv")

# Combine all dataframes into a dictionary for easier management
dfs_dict = {
    'application_categories': AC,
    'application_developers': AD,
    'application_genres': AG,
    'application_platforms': AP,
    'application_publishers': APu,
    'applications': application,
    'categories': categories,
    'developers': developers,
    'genres': genres,
    'platforms': platforms,
    'publishers': publishers,
    'reviews': reviews
}

# Display basic info about each dataset
for name, df in dfs_dict.items():
    print(f"\n{name}:")
    print(f"  Shape: {df.shape}")
    print(f"  Columns: {list(df.columns)}")
```

##### if len(AC) > 0:: This checks if the table (e.g., Categories) actually has data in it. If it’s empty, it skips the merge to avoid errors.
##### suffixes=('', '_cat'): If both tables have columns with the same name (like a "name" column), Pandas needs to distinguish them. This tells Pandas to leave the master column name alone and add _cat to the end of the new column.

```python
# Start with the main applications table as the base
combined_data = application.copy()

# Merge with application-relationship tables (these have appid)
# Merge application_categories
if len(AC) > 0:
    combined_data = combined_data.merge(AC, on='appid', how='left', suffixes=('', '_cat'))

# Merge application_developers
if len(AD) > 0:
    combined_data = combined_data.merge(AD, on='appid', how='left', suffixes=('', '_dev'))

# Merge application_genres
if len(AG) > 0:
    combined_data = combined_data.merge(AG, on='appid', how='left', suffixes=('', '_gen'))

# Merge application_platforms
if len(AP) > 0:
    combined_data = combined_data.merge(AP, on='appid', how='left', suffixes=('', '_plat'))

# Merge application_publishers
if len(APu) > 0:
    combined_data = combined_data.merge(APu, on='appid', how='left', suffixes=('', '_pub'))

print("Combined data shape:", combined_data.shape)
print("\nFirst 5 rows:")
print(combined_data.head())
```
