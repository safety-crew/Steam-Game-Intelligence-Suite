### 1 - Setup Environment and Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib as plt
from pathlib import Path
from IPython.display import display
```

### 2 - Load Raw Data

```python
# Create Path object
raw_dir = Path("data/raw")

# Resolve the raw folder from the notebook working directory or any parent folder.
candidates = [raw_dir] + [Path(p) / "data" / "raw" for p in Path.cwd().resolve().parents]
raw_dir = next((p for p in candidates if p.exists()), None)
if raw_dir is None:
    raise FileNotFoundError(
        "Could not locate the raw data folder. Tried: " + ", ".join(str(p) for p in candidates)
    )

# List of raw files
csv_files = [
    "applications.csv",
    "reviews.csv",
    "genres.csv",
    "categories.csv",
    "developers.csv",
    "platforms.csv",
    "publishers.csv",
    "application_genres.csv",
    "application_categories.csv",
    "application_developers.csv",
    "application_platforms.csv",
    "application_publishers.csv",
]

# Load all files
dfs = {name.replace(".csv", ""): pd.read_csv(raw_dir / name, low_memory=False) for name in csv_files}

# Schema descriptions
schema = {
    "applications": {
        "appid": "Unique app identifier",
        "name": "Game/application title",
        "type": "Application type (e.g. game/software)",
        "is_free": "Whether the app is free",
        "release_date": "Release date",
        "required_age": "Minimum age requirement",
        "short_description": "Summary description text",
        "supported_languages": "Supported languages",
        "header_image": "Header image URL",
        "background": "Background image URL",
        "metacritic_score": "Metacritic score",
        "recommendations_total": "Total recommendations",
        "mat_supports_windows": "Supports Windows",
        "mat_supports_mac": "Supports macOS",
        "mat_supports_linux": "Supports Linux",
        "mat_initial_price": "Initial price",
        "mat_final_price": "Final price",
        "mat_discount_percent": "Discount percentage",
        "mat_currency": "Currency code",
        "mat_achievement_count": "Achievement count",
        "mat_pc_os_min": "Minimum PC OS requirement",
        "mat_pc_processor_min": "Minimum CPU requirement",
        "mat_pc_memory_min": "Minimum RAM requirement",
        "mat_pc_graphics_min": "Minimum GPU requirement",
        "mat_pc_os_rec": "Recommended PC OS",
        "mat_pc_processor_rec": "Recommended CPU",
        "mat_pc_memory_rec": "Recommended RAM",
        "mat_pc_graphics_rec": "Recommended GPU",
        "created_at": "Record creation timestamp",
        "updated_at": "Record update timestamp",
    },
    "reviews": {
        "recommendationid": "Unique review identifier",
        "appid": "App identifier for this review",
        "author_steamid": "Steam ID of review author",
        "author_num_games_owned": "Number of games owned by author",
        "author_num_reviews": "Number of reviews written by author",
        "author_playtime_forever": "Total playtime in minutes",
        "author_playtime_last_two_weeks": "Playtime in last 2 weeks",
        "author_playtime_at_review": "Playtime at time of review",
        "author_last_played": "Last played timestamp",
        "language": "Review language",
        "review_text": "Review text",
        "timestamp_created": "Review creation timestamp",
        "timestamp_updated": "Review update timestamp",
        "voted_up": "Whether the review is positive",
        "votes_up": "Upvotes received",
        "votes_funny": "Funny votes received",
        "weighted_vote_score": "Weighted review score",
        "comment_count": "Number of comments on review",
        "steam_purchase": "Purchased on Steam flag",
        "received_for_free": "Received free copy flag",
        "written_during_early_access": "Written during early access flag",
        "created_at": "Record ingestion creation timestamp",
        "updated_at": "Record ingestion update timestamp",
    },
    "genres": {"id": "Genre identifier", "name": "Genre name"},
    "categories": {"id": "Category identifier", "name": "Category name"},
    "developers": {"id": "Developer identifier", "name": "Developer name"},
    "publishers": {"id": "Publisher identifier", "name": "Publisher name"},
    "platforms": {"id": "Platform identifier", "name": "Platform name"},
    "application_genres": {
        "appid": "App identifier",
        "genre_id": "Genre identifier",
    },
    "application_categories": {
        "appid": "App identifier",
        "category_id": "Category identifier",
    },
    "application_developers": {
        "appid": "App identifier",
        "developer_id": "Developer identifier",
    },
    "application_publishers": {
        "appid": "App identifier",
        "publisher_id": "Publisher identifier",
    },
    "application_platforms": {
        "appid": "App identifier",
        "platform_id": "Platform identifier",
    },
}

# Print summary and display schema as tables
for name, df in dfs.items():
    print(f"\n=== {name}.csv ===")
    print("Shape:", df.shape)
    schema_df = pd.DataFrame(
        [
            {
                "column": col,
                "purpose": schema.get(name, {}).get(col, "No description available"),
            }
            for col in df.columns
        ]
    )
    display(schema_df)
```
